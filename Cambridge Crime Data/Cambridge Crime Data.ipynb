## DATA CLEANING

```python
df['Date of Report'] = pd.to_datetime(df['Date of Report'])
df['date'] = df['Date of Report'].dt.day
df['month'] = df['Date of Report'].dt.month
df['year'] = df['Date of Report'].dt.year
df['time'] = df['Date of Report'].dt.time
```

```python
df.head(5)
```

```python
df=df.drop(columns="Date of Report")
```

```python
df.head()
```

```python
df.isna()
```

```python
df.isna().sum()
```

```python
df= df.fillna("Not Available")
```

```python
df.isna().sum()
```

## EXPLORATORY DATA ANALYSIS

```python
import matplotlib.pyplot as plt
```

```python
years = df['year'].value_counts().sort_index()
plt.bar(years.index, years.values)
plt.title('Crime over the years')
plt.xlabel('Year')
plt.ylabel('Number of Crimes')
plt.show()
```

```python
years = df['Neighborhood'].value_counts().sort_index()
plt.figure(figsize=(10, 8))  
plt.pie(years.values, labels=years.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Crimes in Neighborhood')
plt.axis('equal') 
plt.show()
```

```python
# Finding top 10 crimes
top_10_crimes = df['Crime'].value_counts().head(10)
```

```python
plt.figure(figsize=(10, 6))
top_10_crimes.plot(kind='bar', color='violet')
plt.xlabel('Crime Type')
plt.ylabel('Count')
plt.title('Top 10 Crimes')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#Finding top 10 locations that are high in crimes
top_10_locations = df['Location'].value_counts().head(10)
```

```python
plt.figure(figsize=(10, 6))
bars = plt.barh(np.arange(len(top_10_locations)), top_10_locations.values, color='red')
plt.yticks(np.arange(len(top_10_locations)), top_10_locations.index)
plt.xlabel('Count')
plt.ylabel('Unsafe Locations')
plt.title('Top 10 Unsafe Locations')

# Adding counts on the bars
for i, count in enumerate(top_10_locations.values):
    plt.text(count + 0.1, i, str(count), va='center', ha='left')
    
plt.tight_layout()
plt.show()
```
